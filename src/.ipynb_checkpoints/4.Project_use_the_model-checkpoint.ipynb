```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import transformations as tr
import tables
import climb
import plot_jordi
```

```python
pd.set_option("display.max_columns",None)
```

# Load the model and transformers

```python
# load the model from disk
filename = '../model/finalized_model.sav'
reg = pickle.load(open(filename, 'rb'))
# load transformers
filename2 = '../transformer/cat_enc.sav'
cat_enc = pickle.load(open(filename2, 'rb'))
filename3 = '../transformer/num_tr1.sav'
num_tr_1 = pickle.load(open(filename3, 'rb'))
filename4 = '../transformer/num_tr2.sav'
num_tr_2 = pickle.load(open(filename4, 'rb'))
# conversion table
conversion = tables.conversion_table()
conversion.T
```

# Use the model

Now we can use the model to do some predictions

<b> a) Using my info

```python
X_jordi_row = {'country':'ESP','sex':0,'height':172,'weight':62,'age':32,'years_cl':5}
X_jordi = pd.DataFrame(X_jordi_row,index = [0])
X_jordi
```

Apply transformations

```python
X_jordi_tr = tr.all_transform(X_jordi,cat_enc,num_tr_1,num_tr_2)
```

Use prediction

```python
y_pred_jordi = reg.predict(X_jordi_tr)
```

```python
grade_pred = np.round(y_pred_jordi)[0][0]
conversion.loc[grade_pred,'fra_routes']
```

<b> b) Get yearly grade evolution

```python
country = 'ESP'
sex = 1
height = 167
weight = 53
years_cl = 8
age = 39
X_user = climb.yearly_evolution(country,sex,height,weight,age,years_cl)
```

```python
X_user
```

Apply transformations

```python
X_user_trans = tr.all_transform(X_user,cat_enc,num_tr_1,num_tr_2)
```

```python
X_user_trans
```

Apply predictions

```python
X_user['grade_spec'] = climb.climb_prediction(X_user_trans,reg)
```

```python
X_user = climb.yearly_evolution_trans(X_user,conversion,years_cl)
```

Plot results

```python
plot_jordi.plot_evolution(X_user)
```

This is the result of the espected climbing level during the next 10 years
